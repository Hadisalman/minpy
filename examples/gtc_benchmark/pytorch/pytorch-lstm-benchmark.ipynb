```python
import cPickle as pickle
tft_tables = {'naive' : [], 'sophisticated' : []}
bt_tables = {'naive' : [], 'sophisticated' : []}
ift_tables = {'naive' : [], 'sophisticated' : []}
for N in range(6, 13):
    tft, ift, bt = pickle.load(open('time/naive-lstm-%d' % 2 ** N, 'rb'))
    tft_tables['naive'].append(tft)
    ift_tables['naive'].append(ift)
    bt_tables['naive'].append(bt)
    
    tft, ift, bt = pickle.load(open('time/sophisticated-lstm-%d' % 2 ** N, 'rb'))
    tft_tables['sophisticated'].append(tft)
    ift_tables['sophisticated'].append(ift)
    bt_tables['sophisticated'].append(bt)
```

```python
import matplotlib.pylab as pl
%matplotlib inline
import numpy as np
```

```python
for key, value in tft_tables.items():
    pl.plot(np.vectorize(lambda i : 2 ** i)(np.arange(6, 13)), value, label=key)
pl.legend(framealpha=0, loc=0)
pl.title('forward (training)')
```

```python
for key, value in ift_tables.items():
    pl.plot(np.vectorize(lambda i : 2 ** i)(np.arange(6, 13)), value, label=key)
pl.legend(framealpha=0, loc=0)
pl.title('forward (inference)')
```

```python
for key, value in bt_tables.items():
    pl.plot(np.vectorize(lambda i : 2 ** i)(np.arange(6, 13)), value, label=key)
pl.legend(framealpha=0, loc=0)
pl.title('backward')
```
